```python
import pandas as pd
import csv
```

```python
#Reading the Spam Dataset
st = pd.read_excel (r'D:\Program Files\Project\Spam\Book1.xlsx')
print(len(st.Text))
```

```python
#Reading the messages which consists of both spam and ham messages
newdata=pd.read_excel (r'D:\Program Files\Project\Spam\spam_ham.xlsx')
print(len(newdata.Text))
```

```python
#Finding out which messages are Spam and which are Ham from the message dataset
ck1=[]
ck2=[]
resultlist=[]
spamhamoriginal=[]
spamhamresult=[]
print("Index\tLength of the message\tWords matched with Spam\t\tMacthing Percentage\tType of Message")
for i in range(len(st)):
    ck1.append(st.Text[i].split(" "))
for i in ck1:
    for j in i:
        ck2.append(j)
ck2=set(ck2)        
ck2=list(ck2)
for i in range(len(newdata)):
    cout=0
    if type(newdata.Text[i]) is not int:
        a1=newdata.Text[i].split(" ")
        b1=len(a1)
        for j in range(len(a1)):
            for k in range(len(ck2)):
                if a1[j] == ck2[k]:
                    cout=cout+1
        result=(cout/b1)*100
        result = float(round(result, 2))
        #print(" ",i,"\t\t",b1,"\t\t\t",cout,"\t\t\t     ",result,"%")
        if result > 70:
            print(" ",i,"\t\t",b1,"\t\t\t",cout,"\t\t\t     ",result,"%","\t\t     Spam")
            resultlist.append([i,newdata.Type[i],"spam",result])
            spamhamoriginal.append(newdata.Type[i])
            spamhamresult.append("spam")
        else:
            print(" ",i,"\t\t",b1,"\t\t\t",cout,"\t\t\t     ",result,"%","\t\t     Ham")
            resultlist.append([i,newdata.Type[i],"ham",result])
            spamhamoriginal.append(newdata.Type[i])
            spamhamresult.append("ham")
```

```python
#Finding the Accuracy of the output
co=0
for i in range(len(spamhamoriginal)):
    if spamhamoriginal[i] == spamhamresult[i]:
        co=co+1
accuracy=(co/len(spamhamoriginal))*100
print(accuracy)
```

```python
#Counting total Number of spam and ham messages in the message dataset
c=0
d=0
for i in range(len(newdata)):
    if newdata.Type[i] == "spam":
        c=c+1
    else:
        d=d+1
print("Spam = ",c,"\tHam",d)
```
